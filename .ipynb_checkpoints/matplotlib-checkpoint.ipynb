```python

import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5,6,7]
max_t = [50, 53, 76, 87,76,88, 45]
min_t = [65, 68, 46, 59, 32,27,63]
avg_t = [65, 76, 87, 78, 54,78,98]

plt.title('Weather')
plt.xlabel('Days')
plt.ylabel('Temperature')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend(loc='lower left', shadow=True, fontsize='large')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
branch=['it','ece','cse']
it=[12,30,1,3,22]
ece=[23,6,16,5,10]
cse=[29,3,24,25,17]
a=np.arange(len(it))
barwidth=0.25
plt.bar(a,ece,color='g',width=barwidth,edgecolor='grey',label='ece')
plt.bar(a-barwidth,it,color='r',width=barwidth,edgecolor='grey',label='it')
plt.bar(a+barwidth,cse,color='b',width=barwidth,edgecolor='grey',label='cse')
plt.xlabel('branch')
plt.ylabel('students passed')
plt.legend()
plt.show()
```
