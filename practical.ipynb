```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

# Dataset
data = pd.DataFrame({
    "carprice": [2000000, 4000000, 5000000, 7000000, 9600000, 12000000, 15000000],
    "mileage":  [20,     15,     12,    14,    18,    22,     26]
})

# Now X = mileage, y = price
X = data[['mileage']]
y = data[['carprice']]

# -----------------------------
# 1 Linear Regression
# -----------------------------
model = linear_model.LinearRegression()
model.fit(X, y)

# Predict price for mileage = 18 km/l
linear_pred = model.predict([[18]])
print("Linear Regression Price Prediction:", linear_pred)

# -----------------------------
# 2 Polynomial Regression (degree 2)
# -----------------------------
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model2 = linear_model.LinearRegression()
model2.fit(X_poly, y)

poly_pred = model2.predict(poly.transform([[18]]))
print("Polynomial Regression Price Prediction:", poly_pred)

# -----------------------------
# Plot
# -----------------------------
plt.scatter(X, y)
X_range = np.linspace(min(X.mileage), max(X.mileage), 100).reshape(-1, 1)
plt.plot(X_range, model.predict(X_range))
plt.xlabel("Mileage")
plt.ylabel("Car Price")
plt.title("Linear Regression (Price vs Mileage)")
plt.show()
```

```python
numbers = [10, 20, 4, 45, 99]
first = second = float('-inf')
for n in numbers:
    if n > first:
        second = first
        first = n
    elif n > second and n != first:
        second = n
print("Second largest:", second)
```
