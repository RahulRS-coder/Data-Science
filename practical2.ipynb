```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# --------------------------------------------
# 1. Create DataFrame
# --------------------------------------------

data4 = {
    "Amount (USD)": [50, 200, 500, 1200, 75, 3000, 20, 150, 800, 60,
                     2500, 100, 400, 2200, 35, 700, 90, 1800, 55, 120],
    
    "Transaction_time": [10, 14, 23, 2, 9, 1, 16, 20, 3, 11,
                         0, 13, 18, 4, 15, 21, 8, 5, 22, 19],
    
    "Age": [25, 40, 32, 28, 45, 50, 22, 35, 29, 31,
            48, 27, 36, 42, 24, 33, 30, 46, 26, 38],
    
    "Location": ["New York", "Los Angeles", "Chicago", "Miami", "Dallas",
                 "Houston", "Seattle", "Boston", "San Francisco", "Denver",
                 "Atlanta", "Phoenix", "Detroit", "Orlando", "Portland",
                 "Las Vegas", "San Diego", "Austin", "Columbus", "Charlotte"],
    
    "Fraudulent": ["No", "No", "Yes", "Yes", "No", "Yes", "No", "No", "Yes", "No",
                   "Yes", "No", "No", "Yes", "No", "Yes", "No", "Yes", "No", "No"]
}

df = pd.DataFrame(data4)
print("\nOriginal DataFrame:\n", df)

# --------------------------------------------
# 2. Encode categorical variables
# --------------------------------------------
df['Fraudulent'] = df['Fraudulent'].map({'No': 0, 'Yes': 1})

# One-hot encoding for Location
df_encoded = pd.get_dummies(df, columns=['Location'], drop_first=True)

print("\nEncoded DataFrame:\n", df_encoded)

# --------------------------------------------
# 3. Split into X and y
# --------------------------------------------
X = df_encoded.drop("Fraudulent", axis=1)
y = df_encoded["Fraudulent"]

# --------------------------------------------
# 4. Train-test split (stratified)
# --------------------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42)

# --------------------------------------------
# 5. Logistic Regression Model
# --------------------------------------------
model = LogisticRegression(class_weight="balanced", max_iter=1000)
model.fit(X_train, y_train)

# --------------------------------------------
# 6. Predictions
# --------------------------------------------
y_pred = model.predict(X_test)

# --------------------------------------------
# 7. Evaluation
# --------------------------------------------
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
