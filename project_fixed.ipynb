##### Train random forest model identify top ten important features

```python
from sklearn import datasets
iris=datasets.load_iris()
iris
print(iris.feature_names)
print(iris.target_names)
```

```python
import pandas as pd
data=pd.DataFrame({'sepal length':iris.data[:,0],
                    'sepal width': iris.data[:,1],
                    'petal width':iris.data[:,2],
                     'petal length':iris.data[:,3],
                      'species':iris.target})
data
```

```python
x=data[['sepal length','sepal width','petal width','petal length']]
x
y=data['species']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
x_train, x_test, y_train, y_test
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)
model.fit(x_train, y_train)
```

```python
importances = model.feature_importances_
feature_names = x.columns
```

```python
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": importances
})
```

```python
top10 = importance_df.sort_values(by="Importance", ascending=False).head(10)

print("\nTop 10 Important Features:")
print(top10)
```
